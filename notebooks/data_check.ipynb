```python
def func(text):
    a = text.split(' ')
    for i in a:
        if i[0] == type(int):
            return i


print(func('57000.0 руб./ За месяц, Залог - 60000 руб., Ко...'))
```

```python
def splitter(text):
    a = text.split(',')
    
    return a[0]
```

```python
def intchanger(text):
    res = float(''.join(list(filter(lambda x: x.isdigit() == True, text))[:-1]))
    return (res)

# print(intchanger('Залог - 210000 руб.'))
```

```python
def splitter2(text):
    a = text.split(',')
    return a[1]

# print(splitter2('210000.0 руб./ За месяц, Залог - 210000 руб., ...'))
```

```python
def intchanger2(text):
    try:
        res = int(''.join(list(filter(lambda x: x.isdigit() == True, text))))
        return (res)
    except:
        return None
```

```python
data['Цена'] = data['Цена'].apply(splitter)
data['Цена'] = data['Цена'].apply(intchanger)
```

```python
data['Залог'] = data['Залог'].apply(splitter2)
data['Залог'] = data['Залог'].apply(intchanger2)
```
